# # Data Prétraitement

```python
import pandas as ps
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
titanic=ps.read_csv('titanic.csv')
titanic.info()
titanic.isnull().sum()
titanic.head()
```

```python
titanic
```

```python
l=['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
for i in l :
    titanic[i].fillna(titanic[i].mode(),inplace=True)
```

```python
Sex={'Sex':{'male':1,'female':0}}
titanic.replace(Sex,inplace=True)
```

```python
Embarked={'Embarked':{'C':0,'S':1,'Q':2}}
titanic.replace(Embarked,inplace=True)
```

```python
f=titanic['Cabin'].unique()
Cabin={}
for i in range(len(f)):
    Cabin[f[i]]=i
    Cabin2={'Cabin':Cabin}
```

```python
titanic.replace(Cabin2,inplace=True)
```

```python
titanic['Age'].fillna(titanic['Age'].mean,inplace=True)
```

# # Data visualisation

```python
def pcm( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sb.diverging_palette( 230 , 375, as_cmap = True )

    s = sb.heatmap(corr, cmap = cmap,square=True, cbar_kws={ 'shrink' : .9 },ax=ax,annot = True, annot_kws = { 'fontsize' : 11 })
```

```python
pcm(titanic)
```

```python
f=sb.FacetGrid(titanic,row='Sex',col='Survived')
f.map(sb.barplot,'Embarked','Age',ci=None)
```
